```python
import pandas as pd
import matplotlib.pyplot as plt
zdroje = pd.read_csv('instalovanyVykon.csv')
zdroje
```

```python
zdroje['Průběžný součet'] = zdroje.groupby('Typ')['Výkon(MW)'].cumsum()
zdroje
```

```python
pivot_zdroje = zdroje.pivot(index='Rok', columns='Typ', values='Průběžný součet')
pivot_zdroje = pivot_zdroje.ffill()
pivot_zdroje.head()
```

```python
pivot_zdroje.plot(kind='line', figsize=(18,12))
plt.title('Instalovaný výkon nových obnovitelných zdrojů uvedených do provozu v letech 2013-2022 (MW)')
plt.xlabel('Rok')
plt.ylabel('Instalovaný výkon (MW)')
plt.legend(title='Typ zdroje', loc='best')
```
