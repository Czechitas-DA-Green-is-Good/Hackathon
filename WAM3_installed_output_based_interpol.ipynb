```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
file = "Instalovany_vykon.xlsx"
xls = pd.ExcelFile(file)
xls.sheet_names
```

```python
df = xls.parse("List1")
df = df.rename(columns={"Unnamed: 0": "Rok"})
df.head(10)
```

```python
df_for_interpol = df[0:9]
df_for_interpol = df_for_interpol.set_index("Rok")
df_for_interpol
```

```python
df_for_interpol.index.values.tolist()
```

```python
df_for_interpol.loc["Plynové (PLE)"] = (
    df_for_interpol.loc["Paroplynové (PPE)"]
    + df_for_interpol.loc["Plynové a spalovací (PSE)"]
)
df_for_interpol
```

```python
df_for_interpol = df_for_interpol.drop(
    index=["Instalovaný výkon [MWe]", "Paroplynové (PPE)", "Plynové a spalovací (PSE)"]
)
df_for_interpol
```

```python
df_for_interpol[list(range(2023, 2030))] = float("nan")
df_for_interpol
```

```python
df_for_interpol[2030] = [4290, 3000, 1113.57568, 1171.50000, 1500, 10100, 3200]
df_for_interpol
```

```python
df_interpolated = df_for_interpol.interpolate(method="polynomial", order=2, axis=1)
df_interpolated
```

```python
df_interpolated.transpose().plot(figsize=(10, 8))
```

```python
df_interpolated.to_csv("WAM3 Interpolated Energy Mix.csv", index=True)
```
