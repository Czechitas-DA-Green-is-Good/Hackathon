```python
import pandas as pd

df = pd.read_csv("Interpolated Energy Mix.csv")
df.set_index("Rok", inplace=True)
df
```

```python
df = df / df.sum(axis=0)
```

```python
emmissions_per_kwh = [0.005, 0.9, 0, 0, 0.00464, 0.058, 0.5]
```

```python
df = df.multiply(emmissions_per_kwh, axis=0)
df
```

```python
annual_emmissions = df.sum(axis=0)
annual_emmissions
```

```python
annual_emmissions.plot(y=range(0, 1), title="Emissions per kwh", ylim=[0, 1])
```

```python
annual_emmissions.to_csv("Annual Emmissions Development.csv", index=True)
```
