```python
import pandas as pd
import matplotlib.pyplot as plt

file = "Instalovany_vykon.xlsx"
xls = pd.ExcelFile(file)
xls.sheet_names
```

```python
df = xls.parse("List1")
df = df.rename(columns={"Unnamed: 0": "Rok"})
df.head(20)
```

```python
df_for_interpol = df[9:18]
df_for_interpol = df_for_interpol.set_index("Rok")
df_for_interpol
```

```python
df_for_interpol.index.values.tolist()
```

```python
df_for_interpol.loc["Plynové (PLE)"] = (
    df_for_interpol.loc["Paroplynové (PPE)"]
    + df_for_interpol.loc["Plynové a spalovací (PSE)"]
)
df_for_interpol
```

```python
df_for_interpol = df_for_interpol.drop(
    index=[
        "Výroba elektřiny brutto [MWh]",
        "Paroplynové (PPE)",
        "Plynové a spalovací (PSE)",
    ]
)
df_for_interpol
```

```python
df_for_interpol[list(range(2023, 2030))] = float("nan")
df_for_interpol
```

```python
df_for_interpol[2030] = [
    31315188.94,
    (41017314.5 * 0.5947388761792955),
    2251849.95,
    1100556.4,
    (641330.5 * 2.064318867668019),
    (2298347.9 * 2.856630064840218),
    8783791.5,
]
df_for_interpol
```

```python
df_interpolated = df_for_interpol.interpolate(method="linear", axis=1)
df_interpolated
```

```python
df_interpolated.transpose().plot(figsize=(10, 8))
```
