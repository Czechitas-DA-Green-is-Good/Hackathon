```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

file = "Instalovany_vykon.xlsx"
xls = pd.ExcelFile(file)
xls.sheet_names
```

```python
df = xls.parse("List1")
df = df.rename(columns={"Unnamed: 0": "Rok"})
df.head(20)
```

```python
df_for_interpol = df[9:18]
df_for_interpol = df_for_interpol.set_index("Rok")
df_for_interpol = df_for_interpol / 1000
df_for_interpol
```

```python
df_for_interpol.index.values.tolist()
```

```python
df_for_interpol.loc["Plynové (PLE)"] = (
    df_for_interpol.loc["Paroplynové (PPE)"]
    + df_for_interpol.loc["Plynové a spalovací (PSE)"]
)
df_for_interpol
```

```python
df_for_interpol = df_for_interpol.drop(
    index=[
        "Výroba elektřiny brutto [MWh]",
        "Paroplynové (PPE)",
        "Plynové a spalovací (PSE)",
    ]
)
df_for_interpol
```

```python
df_for_interpol[list(range(2023, 2030))] = float("nan")
df_for_interpol
```

```python
df_for_interpol[2030] = [
    31021.8104399999,
    (41017.3145 * 0.5732388761792955),
    2151.84995,
    1170.4551,
    (641.3305 * 2.065318867668019),
    (2298.3479 * 2.856630064840218),
    9169.9072 * 1.43651644092438,
]
df_for_interpol
```

```python
df_interpolated = df_for_interpol.interpolate(method="polynomial", order=2, axis=1)
df_interpolated
```

```python
df_transposed = df_interpolated.transpose()

colors_dict = {
    "Jaderné (JE)": "#67bf5c",
    "Parní (PE)": "#a8786e",
    "Plynové (PLE)": "#ad8bc9",
    "Vodní (VE)": "#a2a2a2",
    "Přečerpávací (PVE)": "#7f7f7f",
    "Větrné (VTE)": "#ed97ca",
    "Fotovoltaické (FVE)": "#ff9e4a",
}

colors = [colors_dict[column] for column in df_transposed.columns]
ax = df_transposed.plot.area(stacked=True, figsize=(12, 8), color=colors)

plt.title("Výroba elektřiny dle zdrojů s interpolací do r. 2030 (scénář WEM)")
plt.ylabel("Výroba elektřiny podle zdroje (GWh)")
plt.legend(title="Typ elektrárny")

plt.show()
```

```python
df_interpolated.to_csv("Interpolated Production Based Energy Mix.csv", index=True)
```
