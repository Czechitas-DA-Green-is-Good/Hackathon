```python
import pandas as pd
from IPython.display import display

df = pd.read_csv("Elektricke energie_bilance_final.csv", index_col=0)
df = df.rename(
    columns={
        "year": "rok",
        "Vyroba_e": "Celková",
        "p_elektrarny": "Uhelné",
        "vod_elektrarny": "Vodní",
        "vet_elektrarny": "Větrné",
        "sol_elektrarny": "Solární",
        "J_elektrarny": "Jaderné",
    }
)
display(df)
```

```python
file = "Instalovany_vykon.xlsx"
xls = pd.ExcelFile(file)
xls.sheet_names
```

```python
df1 = xls.parse("List1")
df1 = df1.rename(columns={"Unnamed: 0": "Rok"})
df1.head(10)
```

```python
df_interpol = df1[0:9]
df_interpol = df_interpol.set_index("Rok")
df_interpol
```

```python
df_interpol.index.values.tolist()
# ["Paroplynové (PPE)"]  # + df_interpol["Plynové a spalovací (PSE)"]
```

```python
df_interpol.loc["Plynové (PLE)"] = (
    df_interpol.loc["Paroplynové (PPE)"] + df_interpol.loc["Plynové a spalovací (PSE)"]
)
df_interpol
```

```python
df_interpol = df_interpol.drop(
    index=["Instalovaný výkon [MWe]", "Paroplynové (PPE)", "Plynové a spalovací (PSE)"]
)
df_interpol
```

```python
df_interpol[list(range(2023, 2030))] = float("nan")
df_interpol
```

```python
df_interpol[2030] = [4290, 5600, 1113.57568, 1171.50000, 700, 6000, 3800]
df_interpol
```

```python
df_interpolated = df_interpol.interpolate(method="polynomial", order=2, axis=1)
df_interpolated
```

```python
import matplotlib.pyplot as plt
import seaborn as sns


# plt.figsize(12, 9)


df_transposed = df_interpolated.transpose()

colors_dict = {
    "Jaderné (JE)": "#E15759",
    "Parní (PE)": "#9D7660",
    "Plynové (PLE)": "#F1CE63",
    "Vodní (VE)": "#3896c4",
    "Přečerpávací (PVE)": "#a0d4ee",
    "Větrné (VTE)": "#BAB0AC",
    "Fotovoltaické (FVE)": "#B07AA1",
}

colors = [colors_dict[column] for column in df_transposed.columns]
# colors = sns.color_palette("hls", len(df_transposed.columns))
ax = df_transposed.plot.area(stacked=True, figsize=(12, 8), color=colors)

plt.title("Instalovaný výkon dle zdrojů s interpolací do r. 2030 (scénář WEM)")
plt.ylabel("Instalovaný výkon (MWe)")
plt.legend(title="Typ elektrárny")

plt.show()
```

```python
df_interpolated.to_csv("Interpolated Energy Mix.csv", index=True)
```
